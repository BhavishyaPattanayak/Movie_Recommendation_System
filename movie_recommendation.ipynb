```python
df_movies['title'] = df_movies['title'].apply(lambda x: x.strip())
df_movies["title"]=df_movies["title"].apply(lambda x:x.replace(x,x.lower()))
df_movies['title']
```

```python
df_movies["year"] = df_movies["title"].str.extract(r'\((\d{4})\)')
df_movies["title"] = df_movies["title"].str.replace(r'\s*\(\d{4}\)', '')
df_movies
```

```python
df_moviegenre=df_movies.copy()
for index, row in df_movies.iterrows():
    for genre in row['genres']:
        df_moviegenre.at[index, genre] = 1
df_moviegenre= df_moviegenre.fillna(0)
df_moviegenre
```

```python
df_movieratings = pd.read_csv('ratings.csv')
df_movieratings
```

```python
def get_recently_watched_movies():
    new = []
    print("Number of movies watched recently:")
    num_movies = int(input())

    for i in range(num_movies):
        dc = {}
        print("MOVIE WATCHED RECENTLY (in lowercase)=")
        dc["title"] = input()
        print("GIVEABLE RATING (out of 5) =")
        dc["rating"] = float(input())
        new.append(dc)

    usermovies = pd.DataFrame(new)  # Creating the DataFrame
    return usermovies

# Calling the function to get recently watched movies
recently_watched = get_recently_watched_movies()
print("Recently watched movies:")
recently_watched
```
